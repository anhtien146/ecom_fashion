## Data aggregation

```python
df2.columns
```

```python
# Count value on the 'products_sold' column
products_sold = df2['products_sold'].unique()
nb_user = df2['products_sold'].value_counts()
nb_user_sold_prd = pd.DataFrame({'products_sold': products_sold,
                                'nb_user': nb_user})
nb_user_sold_prd.sort_values('products_sold')
```

**Nhận xét:**
- Ta sẽ chia nb_user_sold_prd thành các phần để có thể thấy rõ hơn hành vi của người dùng trong phần sau. Vậy làm sao để chia được thành các phần? Ta sẽ sử dụng Pareto. Với nb_user_sold_prd thì ta sẽ hiểu như sau: 80% lượng hàng được bán đến từ 20% người dùng
- Giá trị người dùng không mua là quá lớn, chiếm hơn 97% nên ta sẽ để giá trị này là 1 phần riêng để phân tính sau

```python
# Calculate the number user percent and cumulative
cond2 = nb_user_sold_prd['products_sold'] > 0
nb_user_sold_prd['nb_user_per'] = nb_user_sold_prd['nb_user'] / nb_user_sold_prd['nb_user'].loc[cond2].sum()
nb_user_sold_prd['cumulative'] = nb_user_sold_prd['nb_user_per'].loc[cond2].cumsum()
```

```python
# Pareto
nb_user_sold_prd[nb_user_sold_prd['cumulative'] > 0.8].sort_values('products_sold')
```

```python
# Count value on the 'products_bought' column
products_bought = df2['products_bought'].unique()
nb_user_bought = df2['products_bought'].value_counts()
nb_user_bought_prd = pd.DataFrame({'products_bought': products_bought,
                                'nb_user_bought': nb_user_bought})
nb_user_bought_prd.sort_values('products_bought')
```

```python
# Calculate the number user percent and cumulative
cond3 = nb_user_bought_prd['products_bought'] > 0
nb_user_bought_prd['nb_user_per'] = nb_user_bought_prd['nb_user_bought'] / nb_user_bought_prd['nb_user_bought'].loc[cond3].sum()
nb_user_bought_prd['cummalative'] = nb_user_bought_prd['nb_user_per'].loc[cond3].cumsum()
nb_user_bought_prd[nb_user_bought_prd['cummalative'] > 0.8].sort_values('products_bought')
```

**Tổng kết:**
- Với 'product_sold' ta sẽ chia thành các phần như sau:
    + Không mua: 'product_sold' = 0
    + Mua ít: 0 < 'product_sold' < 6
    + Mua nhiều: 6 <= 'product_sold'
- Với 'product_bought' ta sẽ chia thành các phần như sau:
    + Không mua: 'product_sold' = 0
    + Mua ít: 0 < 'product_sold' < 4
    + Mua nhiều: 4 <= 'product_sold'

-> Các giá trị được chia chỉ phù hợp với bộ dữ liệu này
